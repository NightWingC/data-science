```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'Nombre': ["Juan", "Gabriela", "Elena"], 'Edad': [23, 31, 21]})
df1
```

```python
df2 = pd.DataFrame({'Nombre': ["Carmela", "Max", "Laura"], 'Edad': [34, 25, 29]})
df2
```

```python
df_concatenar = pd.concat([df1,df2])
df_concatenar
```

```python
df_concatenar2 = pd.concat([df1,df2], axis=1)
df_concatenar2
```

```python
df_concatenar = pd.concat([df1,df2], ignore_index=True)
df_concatenar
```

```python
df_concatenar = pd.concat([df1,df2], keys=["df1","df2"])
df_concatenar
```

```python
df_concatenar = pd.concat([df1,df2], keys=["Enero","Febrero"])
df_concatenar
```

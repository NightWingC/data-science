```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'ID': [1, 2, 3],
                    'Nombre': ['Ana', 'Luis', 'Carlos']})
df2 = pd.DataFrame({'ID': [1, 2, 4],
                    'Edad': [25, 30, 22]})
```

```python
df1
```

```python
df2
```

```python
df_combinado = pd.merge(df1,df2, on=("ID"), how="outer")
df_combinado
```

```python
df_indexado = pd.merge(df1,df2, left_index=True, right_index=True)
df_indexado
```

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'ID': [1, 2, 3],
                    'Nombre': ['Ana', 'Luis', 'Carlos']})
df2 = pd.DataFrame({'ID': [1, 2, 4],
                    'Edad': [25, 30, 22]})
```

```python
df1
```

```python
df2
```

```python
df_combinado = pd.merge(df1,df2, on=("ID"), how="outer")
df_combinado
```

```python
df_indexado = pd.merge(df1,df2, left_index=True, right_index=True)
df_indexado
```

```python
libros = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'titulo': ['El Quijote', 'La Odisea', 'Cien Aos de Soledad', 'El Principito']
})

autores = pd.DataFrame({
    'ID': [1, 2, 3, 5],
    'nombre': ['Miguel de Cervantes', 'Homero', 'Gabriel Garca Mrquez', 'J.K. Rowling']
})

# Fusionando los DataFrames usando un outer join
libros_autores = pd.merge(libros,autores, on="ID", how="outer")
libros_autores
```

```python
cursos = pd.DataFrame({
    'curso_id': [101, 102, 103],
    'nombre_curso': ['Introduccin a Python', 'Data Science con Python', 'Machine Learning Avanzado']
})

inscripciones = pd.DataFrame({
    'curso_id': [102, 102, 101, 104],
    'fecha_inscripcion': ['2023-01-15', '2023-02-01', '2023-01-20', '2023-03-05']
})

# Fusionando cursos con inscripciones para incluir todos los registros de inscripciones
cursos_inscripciones = pd.merge(cursos,inscripciones, on="curso_id", how="left")
cursos_inscripciones
```

```python
productos = pd.DataFrame({
    'ID': [10, 11, 12],
    'Nombre': ['Teclado', 'Mouse', 'Monitor'],
    'Marca': ['Logitech', 'Razer', 'Dell']
})

reviews = pd.DataFrame({
    'ID': [10, 11, 12],
    'Calificacin': [5, 4, 4],
    'Comentario': ['Excelente producto', 'Buen producto', 'Satisfecho']
})

# Fusionando productos con reviews manteniendo los ndices originales
productos_reviews = pd.merge(productos,reviews, left_index=True, right_index=True)
productos_reviews
```

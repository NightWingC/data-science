```python
import pandas as pd
```

```python
ruta = 'https://raw.githubusercontent.com/NightWingC/data-science/refs/heads/main/CodigoPython/csv/Top_Peliculas.csv'
df = pd.read_csv(ruta)
```

```python
df.head(10)
```

```python
df_ordenado = df.sort_values(by="rating")
df_ordenado.head(6)
```

```python
df_ordenado = df.sort_values(by="rating", ascending=False)
df_ordenado.head(6)
```

```python
df_ordenado = df.sort_values(by=["rating", "recaudación(M)"], ascending=False)
df_ordenado.head(6)
```

```python
df_agrupado = df.groupby("género")["rating"].mean()
df_agrupado
```

```python
df_agrupado = df.groupby("año")["recaudación(M)"].sum()
df_agrupado.sort_values(ascending=False).head(10)
```

```python

datos = {
    'titulo': ['Pelicula A', 'Pelicula B', 'Pelicula C', 'Pelicula D'],
    'año': [2001, 2000, 2005, 2010],
    'rating': [7.2, 6.5, 8.1, 7.5]
}

df_ejercicio = pd.DataFrame(datos)
df_ejercicio
```

```python
# Ordenando el DataFrame por la columna 'año' en orden ascendente
df_ordenado = df_ejercicio.sort_values("año", ascending=False)
df_ordenado
```

```python
# Ordenar las películas por 'rating' descendente y luego por 'año' ascendente
df_ordenado2 = df_ejercicio.sort_values(by=["rating", "año"], ascending=[False,True])
df_ordenado2
```

```python
datos2 = {
    'titulo': ['Pelicula A', 'Pelicula B', 'Pelicula C', 'Pelicula D', 'Pelicula E'],
    'género': ['Acción', 'Drama', 'Acción', 'Comedia', 'Drama'],
    'rating': [7.2, 8.5, 9.1, 6.5, 7.8]
}

df3 = pd.DataFrame(datos2)
df3
```

```python
# Agrupar películas por 'género' y calcular el promedio de 'rating' para cada género
promedio_rating_por_genero = df3.groupby("género")["rating"].mean()
promedio_rating_por_genero
```

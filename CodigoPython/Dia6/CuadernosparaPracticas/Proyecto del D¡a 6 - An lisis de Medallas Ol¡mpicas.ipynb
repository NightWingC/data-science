```python
# Reemplazar valores no validos
df.fillna(0, inplace=True)
df
```

```python
df["Oro"] = df["Oro"].astype(int)
df["Plata"] = df["Plata"].astype(int)
df["Bronce"] = df["Bronce"].astype(int)
df
```

```python
# Analisis de medallas de oro por país
top_3_oro = df.sort_values("Oro", ascending=False).head(3)
top_3_oro
```

```python
# Análisis de medallas totales por país
filtro = df["Total"] > 10
mas_10_medallas = df[filtro]
mas_10_medallas.sort_values("Total", ascending=False)
```
